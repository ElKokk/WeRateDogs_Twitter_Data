```python
import tweepy
from tweepy import OAuthHandler
import json
from timeit import default_timer as timer

# Query Twitter API for each tweet in the Twitter archive and save JSON in a text file
# These are hidden to comply with Twitter's API terms and conditions
consumer_key = 'HIDDEN'
consumer_secret = 'HIDDEN'
access_token = 'HIDDEN'
access_secret = 'HIDDEN'

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True)


tweet_ids = df.tweet_id.values
len(tweet_ids)

# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

## Clean - Quality Issue #5
The numerators and denominators of the ratings include mistakes and should be fixed. The denominators above 10 do not include ratings.

**Define**

We will extract the numerators and denominators from the text, using a regular expression. We will convert them to gloat and drop the rows that have denominator different than 10.
Then we will do the division and create a new column with the rating of the dog. Finally we will drop the outliers that may not reflect actual ratings and then we will drop the numerator and denominator columns as we do not need them anymore.

**Code**

```python
#Reference Regular Expression: https://knowledge.udacity.com/questions/48718

nd=tweet_info_clean.text.str.extract('((?:\d+\.)?\d+)\/(\d+)',expand=True)
nd[0]=nd[0].astype(float)
nd[1]=nd[1].astype(float)
nd=nd[nd[1]==10]   #Picking only denominators ==10

tweet_info_clean['rating']=nd[0]/nd[1]
```

```python
#Dropping numerator and denominator columns

tweet_info_clean.drop(['rating_numerator','rating_denominator'],axis=1,inplace=True)
```

## Clean - Tidiness Issue #2

The columns `doggo`,`floofer`,`pupper`,`puppo`, are all different categorical values of the same variable that is the dog "stage". So they should be merged into one column.

**Define**

First will replace the `None` values in each stage column with empty strings. Then we will add each stage in a new column called 'stage'. Since there are some rows with two stages, we will locate them and separate them. Finally we will replace the empty strings with NaN values, we will drop the individual stage columns and we will convert stage to a categorical type.

```python
tweet_info_clean.puppo.value_counts()
```

```python
#We will first transform the stages columns back to strings
```

```python
stages=['doggo','floofer','pupper','puppo']

for stage in stages:
    tweet_info_clean[stage]=tweet_info_clean[stage].astype(str)
    tweet_info_clean[stage]=tweet_info_clean[stage].replace('None','')
```

```python
tweet_info_clean['stage']=tweet_info_clean['doggo']+tweet_info_clean['floofer'] \
                                                                        +tweet_info_clean['pupper']+ \
                                                                        tweet_info_clean['puppo']
```

```python
#For a few observations, multiple stages exist.

tweet_info_clean.stage.value_counts()
```

```python
tweet_info_clean.loc[tweet_info_clean.stage=='doggopupper','stage']='doggo, pupper'
tweet_info_clean.loc[tweet_info_clean.stage=='doggopuppo','stage']='doggo, puppo'
tweet_info_clean.loc[tweet_info_clean.stage=='doggofloofer','stage']='doggo, floofer'
tweet_info_clean.loc[tweet_info_clean.stage=='nannannannan','stage']=np.nan
```

```python
#We will also replace the empty strings with nan values

tweet_info_clean.replace('',np.nan,inplace=True)
```

```python
#We will now drop the individual stage columns as we do not need them anymore

tweet_info_clean.drop(stages,axis=1,inplace=True)
```

```python
#Finally we will turnt he stage values back to their categorical form

tweet_info_clean.stage=tweet_info_clean.stage.astype('category')
```
